# 📌 Agent 압축 개요

---

## 1. Agent란?
- **정의**:  
  LLM이 단순 답변을 넘어서 **스스로 판단 → 도구 실행 → 결과 활용**하는 구조

- **핵심 요소 4가지**
  1. **목표(Goal/Task)**: 사용자의 요청
  2. **정책(Reasoner/Policy)**: 어떤 행동을 할지 결정
  3. **도구(Tools)**: 계산기, 검색기, DB, 코드 실행 등
  4. **메모리(State/History)**: 대화와 실행 로그 기록

---

## 2. 기본 루프 (Agent Loop)
- **질문 → 생각 → 도구 실행 → 관찰 → 반복 → 최종 답변**

→ 이 과정을 **에이전트 루프**라고 부름.

---

## 3. 간단 예시
**사용자:** “3+5 계산해줘”  
**에이전트 사고:** “계산기가 필요하다”  
**도구 실행:** 계산기 → 8  
**최종 답변:** “답은 8입니다”

---

## 4. LangGraph와의 연결
- 에이전트는 기본적으로 **보이지 않는 루프**
- **LangGraph = 이 루프를 그래프로 시각화하고 제어하는 프레임워크**
  - **Node(노드)** = 생각 / 도구 실행 / 판단
  - **Edge(엣지)** = 다음 단계로 이동 규칙
  - **State(상태)** = 현재 맥락 (질문, 도구 결과, 로그 등)
  - **Interrupt/Checkpoint** = 중단·재개, 상태 저장 지원

---

#### TypedDict
- dict와의 차이점
  - 타입검사 제공 (경고메시지를 출력)
  - 각 키에 대한 구체적인 타입을 지정
  - 생성시 정의한 구조에 따라 사용해야한다 (추가적인 키는 타입 오류가 발생)
- TypedDict 를 사용하는 이유
  - 엄격한 타입 검사로 잠재적인 버그 방지
  - 딕셔너리 구조를 활용하기 때문에 명확하게 정의할 수 있어 코드 가독성 향상

# dict vs TypedDict 비교

| 구분 | dict | TypedDict |
|------|------|-----------|
| **정의 방식** | 키와 값을 자유롭게 정의 | 키와 값의 **타입을 미리 명시** |
| **타입 검사** | 없음 → 잘못된 타입 넣어도 실행됨 | **엄격한 타입 검사** 제공 → 잘못된 타입 시 오류 발생 |
| **유연성** | 자유롭게 키 추가/삭제 가능 | 정의된 구조 안에서만 사용 가능 (추가 키 오류) |
| **가독성** | 구조 파악 어려움 | **IDE 자동 완성, 타입 힌트** 제공 → 코드 가독성↑ |
| **안정성** | 실수로 잘못된 값 넣어도 실행됨 (런타임 에러 가능) | **컴파일 단계에서 버그 예방** 가능 |
| **교육/실습** | 빠르게 테스트할 때 적합 | **실무/협업 코드**에 적합, 구조 명확 |

✅ 정리:  
- **dict** → 빠르게 실습할 때 유용  
- **TypedDict** → 구조를 명확히 정의하고, 버그 방지와 협업 시 안정성 확보에 유리

### Annotated
- 타입힌트에 추가적인 정보 ('메타데이터')를 포함하는 기능
  <pre>
name: Annotated[str, "이름"]
age: Annotated[int, "나이"]
</pre>  

- 주요기능
  - 추가정보를 제공
  - LangGraph 에서 Anotated를 사용하여 특별한 동작 정의
  - 각 필드에 조건/설명/제약을 붙여서 검증가능

```python
from typing import Annotated
from pydantic import Field, BaseModel, ValidationError
# Pydantic : 데이터 유효성 검사 및 관리 라이브러리 -> 타입힌트를 실제로 강제
# ValidationError: 유효성 검증에서 실패했을때 발생하는 예외클래스
# Field: 설정도구 -> 최소길이, 최대길이, 값의 범위 등을 지정

class Member(BaseModel):
  name : Annotated[str, Field(min_length=2, max_length=16, discription = '이름')]
  age : Annotated[int, Field(gt = 18, lt = 65, discription = '나이 (19-64세)')] # 19 ~ 64
```

###  Field 속성 정리표

##### 문자열 전용
| 속성           | 설명        | 예시                              |
| ------------ | --------- | ------------------------------- |
| `min_length` | 최소 길이     | `Field(min_length=3)`           |
| `max_length` | 최대 길이     | `Field(max_length=20)`          |
| `pattern`    | 정규식 패턴 검사 | `Field(pattern="^[a-z0-9_]+$")` |


##### 숫자 전용

```python
#LangGraph 에서 사용할 상태 (State) 설계도
class GraphState2 (TypedDict):
  context : str
  question : str
  answer : str
  score : str
  relevance: str
```

##### Node (노드)
- 함수로 정의
  - 입력, 실행문, 출력
    - 입력: State객체
    - 실행문: 내부 실행할 코드
    - 출력: 반환 return -> 대부분 State객체임

```python
# retrieve node 정의
def retrieve_Node(state: GraphState2) -> GraphState2:
  # Question에 대한 검색 문서를 retriever로 수행
  #retrieve_docs = 검색기이름.invoke(state["question"])
  retrieve_docs = "문서1" # 단순샘플을 입력

  # 검색한 문서를 context 키에 저장하여 반환
  return GraphState2(context=retrieve_docs)

# llm_answer node 정의
def llm_answer_Node(state: GraphState2) -> GraphState2:
    state['answer'] = '답변1'
    return state

# relevance_check node 정의
def relevance_check_Node(state: GraphState2) -> GraphState2:
    state['relevance'] = 'grounded' # 관련이 있음 -> 예시데이터를 받았다고 가정
    return state
```

- graph 에 노드를 추가
  - add_node("노드이름", 노드함수명)

```python
!pip install -q langgraph
```

```python
from langgraph.graph import StateGraph

# 그래프 생성
graph_builder = StateGraph(GraphState2)

# 노드 추가
graph_builder.add_node('retrieve', retrieve_Node)
graph_builder.add_node('llm_answer', llm_answer_Node)
graph_builder.add_node('relevance', relevance_check_Node)

# 입력은 그래프 실행 초기에 직접 State를 넣어줄 예정!
```

##### Edge (엣지)
- 노드와 노드간 연결
- add_edge("시작노드명","다음노드명")

```python
graph_builder.add_edge('retrieve', 'llm_answer')
graph_builder.add_edge('llm_answer','relevance')
```

##### Conditaional Edge (조건부 엣지)
- 노드에 조건부 엣지를 추가하여 상황에 따른 경로를 설정
- add_conditional_edges(source, path, path_map)
  - source : 조건 검사를 시작할 노드
  - path: 현재 상태를 평가하고 다음노드를 결정하는 함수
  - path_map: 조건부 엣지에 연결된 다음 노드 목록  (딕셔너리형태로 작성)

```python
from langgraph.graph import END
# 조건
# grounded (관련이 있다) -> 종료
# NotGrounded (관련이 없다) -> 재검색 실행 ('retrieve')
# notSure (모호하다) -> 다시 답변을 생성 ('llm_answer')

# 관련성이 있는지 확인하는 함수
def is_relevance (state: GraphState2):
  return state['relevance']

# relevance노드에 조건부 엣지를 추가
graph_builder.add_conditional_edges(
    source = "relevance", # 조건을 시작할 노드 이름
    path = is_relevance,
    path_map = {'grounded': END, #  관련성이 있으면 END 끝내세요
                'notGrounded': 'retrieve', # 관련성이 없으면 재검색하세요
                'notSure': 'llm_answer'} # 결과가 모호하면 다시 답변 생성하세요
)
```

### LangGraph를 활용한 챗봇 구축
- langgraph를 활용하여 간단한 챗봇 구현
  - 상태(state)정의
  - 노드(node)정의
  - 그래프 생성 및 노드 추가
  - 엣지(egde) 추가 -> 노드들끼리 어떻게 연결되는지 지정
  - 그래프 컴파일
  - 그래프 시각화
  - 그래프 실행

- 목표 : 메시지를 입력 받아 -> llm 호출 -> 응답 messages 추가 -> 종료

```python
#1. 상태 정의
from typing import Annotated, TypedDict
from langgraph.graph.message import add_messages #리듀서 함수 (reducer)
# add_messages : langgraph에서 제공하는 메세지용 병합 규칙 함수 
# state 내에서 특정키 (messages)를 업데이트 할 때 기본 덮어쓰기 대신 리스트 형태로 이어 붙이기로 처리
class State(TypedDict):
    messages : Annotated[list, add_messages]
```

```python
#2. Node 정의
from langchain_openai import ChatOpenAI
# llm 정의
llm = ChatOpenAI(model='gpt-4o-mini', temperature=0, max_tokens=500)
# chatbot node 구성
def chatbot(state: State):
    return {'messages': [llm.invoke(state['messages'])]}
```

```python
#3. 그래프 생성 및 노드 추가 
from langgraph.graph import StateGraph

# 객체 생성
graph_builder = StateGraph(State)
# 노드 추가
graph_builder.add_node('chatbot', chatbot) #노드 이름, 함수이름
```

```python
#4. 엣지 추가
from langgraph.graph import START, END
# 시작점 -> 챗봇
graph_builder.add_edge(START, 'chatbot') 
# 챗봇 -> 끝
graph_builder.add_edge('chatbot', END) 
```

| 구분       | `add_edge(START, "chatbot")` | `set_entry_point("retrieve")`         |
| -------- | ---------------------------- | ------------------------------------- |
| 동작       | 사용자가 직접 START → 노드 엣지 추가     | 내부적으로 START → 해당 노드 엣지를 자동 추가         |
| 시작 노드 개수 | 여러 개 가능                      | 하나만 지정 가능                             |
| 직관성      | 분기나 다중 시작점 제어에 유리            | 단일 진입점일 때 코드 간결                       |

```python
#5. 그래프 컴파일
graph_chatbot = graph_builder.compile()
```

```python
#6. 그래프 시각화
from langchain_teddynote.graphs import visualize_graph

visualize_graph(graph_chatbot)
```

```python
# 초기 질문 (state의 초기값으로 들어감)
question = '서울의 유명한 맛집 TOP3 추천해주세요'

# 그래프 실행
for res in graph_chatbot.stream({'messages':[('user',question)]}):
    for value in res.values():
        print(value)
        print(value['messages'][-1].content)
```

### langgraph를 활용한 Agent 구축
- llm 도구를 바인딩하여 LLM 입력 된 요청에 따라 필요 시 웹 검색 도구를 호출하는 Agent 구축
- 조건부 Edge를 활용하여 도구 호출 여부에 따라 다른 노드로 전달하는 방법

```python
# 1. state 정의
class State(TypedDict):
    messages : Annotated[list, add_messages]
```

```python
# 2. 웹 검색도구 정의
from langchain_teddynote.tools.tavily import TavilySearch
# 검색도구 실행
tool_tavily = TavilySearch(max_results=2) # Tavily 검색 인스턴스 생성(최대 2건 반환)
# 도구 목록에 추가
tools = [tool_tavily] # llm에 바인딩할 도구 리스트 
```

- llm 을 정의하고 도구를 바인딩
    - bind_tools() : 도구들을 바인딩하는 함수

1. 원래 기본 llm 단방향 (입력 -> 출력)으로 텍스트만 생성
- 즉 사용자의 입력이 들어오면 답변 문자열을 출력 할 뿐 "검색해볼까?, 계산기 돌려볼까?" 도구를 호출하는 로직은 없다!
2. 도구 바인딩을 통해서 API를 호출하여 결과값을 출력 

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model = 'gpt-4o-mini', temperature=0, max_tokens = 500)
# 웹겁색 도구를 바인딩
llm_with_tools = llm.bind_tools(tools)
```

```python
# 챗봇노드 정의
 # 함수 이름의 매개변수 뒤에는 typeddict 를 말하는 것 


def chatbot(state: State):
    return {"messages": [llm_with_tools.invoke(state["messages"])]}
```

```python
# 그래프생성 및 노드 추가
from langgraph.graph import StateGraph



# 객체 생성
graph_builder = StateGraph(State)
# 노드 추가
graph_builder.add_node('chatbot', chatbot) #노드 이름, 함수이름
```

```python
from langgraph.prebuilt import ToolNode # 랭그래프 도구 전용 노드정의 도구
# 도구 노드 정의 및 추가
tool_node = ToolNode(tools)
graph_builder.add_node('tools',tool_node)
```

- 조건부 엣지

```python
from langgraph.prebuilt import tools_condition
# 마지막의 AI 메시지에 도구를 호출하라 의도를 확인하면 -> tools로 이동
graph_builder.add_conditional_edges('chatbot', tools_condition)
```

```python
# 엣지추가
# 시작 -> chatbot
graph_builder.add_edge(START,'chatbot')
# tools -> chatbot
graph_builder.add_edge('tools','chatbot')

# end로 보내지 설정하지 않는 이유?
# tools_condition -> 의도를 확인하여 'tool_calls'라는 메시지가 없으면 알아서 end 보내줌
```

```python
# 그래프 컴파일
graph = graph_builder.compile()
```

```python
# 그래프 시각화
visualize_graph(graph)
```

### Agent에 메모리추가하기
- 단순 llm은 이번 데이터에 대한 상호작용이 없어 스스로를 기억 할 수 없음
- memory 기능을 추가하여 문제 해결
    - checkpointer : 각 노드간의 실행 결과를 추적하기 위한 메모리
    - 그래프를 compile 할 때 checkpointer를 제공하고 그래프 호출 시 thread_id를 제공하여 사용자도 구분함.


```python
%%writefile ./module/LangraphTavilySearch.py
from typing import Annotated, TypedDict  # 타입 힌트 위해 Annotated, TypedDict
from langchain_teddynote.tools.tavily import TavilySearch  # 웹검색 도구(Tavily) 래퍼 임포트.
from langchain_openai import ChatOpenAI
from langgraph.graph import StateGraph, START, END  # 그래프 빌더(StateGraph)와 특수 노드 토큰(START/END)
from langgraph.graph.message import add_messages  # 리듀서(reducer): messages 이어붙이기 규칙
from langgraph.prebuilt import ToolNode, tools_condition
# ToolNode: LLM이 요구한 tool_calls를 실제로 실행하는 노드
# tools_condition: 마지막 AI 메시지에 tool_calls가 있으면 tools로, 없으면 END로 라우팅


# 바깥에서 불러쓸 팩토리 함수. 그래프 빌더를 구성해 반환.
def execute():
    ########## 1. 상태 정의 ##########
    class State(TypedDict):
        messages: Annotated[list, add_messages]

    ########## 2. 도구 정의 및 바인딩 ##########
    tool1 = TavilySearch(max_results=3)  # Tavily 검색 도구
    tools = [tool1]

    llm = ChatOpenAI(model="gpt-4o-mini")
    llm_with_tools = llm.bind_tools(tools)

    ########## 3. 노드 정의 ##########
    def chatbot(state: State):
        return {"messages": [llm_with_tools.invoke(state["messages"])]}

    tool_node = ToolNode(tools=tools)

    ########## 4. 그래프 생성 및 노드 추가 ##########
    graph_builder = StateGraph(State)
    graph_builder.add_node("chatbot", chatbot)
    graph_builder.add_node("tools", tool_node)

    ########## 5. 엣지 추가 ##########
    graph_builder.add_edge(START, "chatbot")
    graph_builder.add_edge("tools", "chatbot")
    graph_builder.add_conditional_edges("chatbot", tools_condition)

    ########## 6. 그래프 빌더 반환 ##########
    return graph_builder
```

```python
# 저장한 모듈에서 execute 로딩
from module.LangraphTavilySearch import execute

graph_builder3 = execute()
```

```python
# 메모리 저장소 생성
from langgraph.checkpoint.memory import MemorySaver
memory = MemorySaver()
```

```python
# 그래프 컴파일
graph3 = graph_builder3.compile(checkpointer=memory)
```

```python
from langchain_core.runnables import RunnableConfig
# RunnableConfig 설정
config = RunnableConfig(recursion_limit = 10, configurable = {'thread_id':'user-1'}) # 실행세션을 구분하는 ID
```

```python
# 실행
# 첫번째 질문
question = '안녕하세요 제 이름은 유동훈입니다 저는 인공지능을 배우는 학생입니다.'

# steam 출력 방식, runnableconfig 추가
for res in graph3.stream({'messages':[('user',question)]},config=config):
    for value in res.values():
        value['messages'][-1].pretty_print()
```
